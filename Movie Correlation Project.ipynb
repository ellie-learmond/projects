```python
# Import packages to use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in data

df = pd.read_csv('/Users/ellielearmond/Downloads/movies.csv')
```

```python
# Data check
df.head()
```

```python
# Checking for any null values
df.isnull().sum()
```

```python
# Dropping null values
df = df.dropna()
```

```python
# Verfiying rows with null values have been dropped
df.isnull().sum()
```

```python
df.head()
```

```python
# Column data types

df.dtypes
```

```python
# Coverting some column data types

df['votes'] = df['votes'].astype('int64')

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Extracting correct release date and adding new column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
# Order data by gross

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.head()
```

```python
# Scatter plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'red'}, line_kws={'color':'green'})
```

```python
# Looking at correlation

df.corr()
```

```python
# 
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix of Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Correlation between gross and rating

sns.barplot(x ="rating", y="gross", data=df)
```

```python
# Looking at the top 10 companies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:10]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
